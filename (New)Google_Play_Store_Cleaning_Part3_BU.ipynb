```python
#create a function to return skewness of all num cols
def calc_skewness(df, num_cols):
    skewness_list = []


    #calculate the skewness
    for col in num_cols:
        skewness = df[col].skew()
        skewness_list.append({"Feature": col, 'Skewness': skewness})

    skewness_df = pd.DataFrame(skewness_list)


    # return the data frame and sort by absolute value of skewness
    return skewness_df.sort_values(by = 'Skewness', key = abs, ascending = False)
```

```python
skewness_df = calc_skewness(df, nums)
```

```python
def conv_to_cat(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            df[col]=df[col].astype('category')
    return df
```

```python
from sklearn.impute import SimpleImputer

def impute_vals(df, num_cols, cat_cols):
    # Numeric imputer
    num_imp = SimpleImputer(strategy='median')
    df[num_cols] = num_imp.fit_transform(df[num_cols])

    # Categorical imputer
    cat_imp = SimpleImputer(strategy='most_frequent')
    df[cat_cols] = cat_imp.fit_transform(df[cat_cols])

    return df
```

```python
# Apply the function to your dataframe
df = impute_vals(df, nums, cat)
```
